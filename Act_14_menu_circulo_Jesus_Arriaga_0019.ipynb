#### 📝 Redacción del problema
* En el estudio de la geometría básica, es común necesitar calcular propiedades fundamentales de figuras circulares y esféricas a partir de una única medida:  * el radio. Para resolver estos cálculos de manera rápida y reutilizable, se * puede automatizar el proceso mediante un programa en Python. La herramienta deberá mostrar un menú interactivo que permita al usuario elegir entre tres
* operaciones: calcular el área de un círculo, calcular su perímetro (circunferencia) o calcular el volumen de una esfera. Cada vez que se seleccione una opción, el programa solicitará el valor del radio, aplicará la
* fórmula matemática correspondiente y mostrará el resultado en pantalla. El flujo de ejecución se mantendrá activo gracias a un ciclo que permita realizar múltiples cálculos hasta que el usuario decida finalizar la sesión.

Fórmulas a utilizar
1. Área del círculo: A=π⋅r 2
2. Perímetro del círculo: P=2⋅π⋅r
3. Volumen de la esfera: V= (3/4)⋅π⋅r 3

```python
# -----------------------------------
#   PROGRAMA DE CALCULOS GEOMETRICOS
# ELABORADO POR JESUS ARRIAGA
# formulas utilizadas
# 1. Área del círculo: A=π⋅r 2
# 2. Perímetro del círculo: P=2⋅π⋅r
# 3. Volumen de la esfera: V= (3/4)⋅π⋅r 3
# -----------------------------------

import math

def areacirculo():
    print("\n📐 Cálculo del Área del Círculo")
    radio = float(input("📏 Ingresa el radio del círculo: "))
    area = math.pi * radio ** 2
    print(f"✅ El área del círculo es: {area:.2f} unidades²\n")

def perimetrocirculo():
    print("\n📏 Cálculo del Perímetro del Círculo")
    radio = float(input("📏 Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"✅ El perímetro del círculo es: {perimetro:.2f} unidades\n")

def volumenesfera():
    print("\n🌍 Cálculo del Volumen de la Esfera")
    radio = float(input("📏 Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * radio ** 3
    print(f"✅ El volumen de la esfera es: {volumen:.2f} unidades³\n")

def menu():
    print("🌟 Bienvenido al Calculador Geométrico 🌟")
    while True:
        print("\n📋 MENÚ PRINCIPAL:")
        print("1️⃣  Calcular área del círculo")
        print("2️⃣  Calcular perímetro del círculo")
        print("3️⃣  Calcular volumen de la esfera")
        print("4️⃣  Salir")

        opcion = input("👉 Selecciona una opción (1-4): ")

        if opcion == "1":
            areacirculo()
        elif opcion == "2":
            perimetrocirculo()
        elif opcion == "3":
            volumenesfera()
        elif opcion == "4":
            print("👋 ¡Gracias por usar el programa! Hasta pronto. 🚀")
            break
        else:
            print("⚠️ Opción no válida. Por favor, selecciona un número del 1 al 4.\n")

# Ejecución principal del programa
if __name__ == "__main__":
    menu()
```
